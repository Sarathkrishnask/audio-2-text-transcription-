```python
import soundfile as sf
import numpy as np
from simple_diarizer.diarizer import Diarizer
ls =[]

# Load the audio
audio_path = "/content/case_study.wav"
signal, fs = sf.read(audio_path)

# Run diarization
diar = Diarizer(embed_model='xvec', cluster_method='sc')
segments = diar.diarize(audio_path, num_speakers=2)

# Sort segments by start time (chronological order)
segments = sorted(segments, key=lambda x: x["start"])

# Split and save each turn
for i, seg in enumerate(segments, 1):
    speaker = seg["label"]
    start_sample = int(seg["start"] * fs)
    end_sample = int(seg["end"] * fs)
    chunk = signal[start_sample:end_sample]

    filename = f"turn_{i:02d}_speaker_{speaker}.wav"
    sf.write(filename, chunk, fs)
    # ls.append(filename)
    # print(f"Saved: {filename} — Speaker {speaker}, Duration: {seg['end'] - seg['start']:.2f}s")
    # Calculate duration in seconds
    start_time = round(start_sample / fs, 2)
    end_time = round(end_sample / fs, 2)

    # Append structured info
    ls.append({
        "speaker": f"speaker_{speaker}",
        "filename": filename,
        "start": start_time,
        "end": end_time
    })

    # print(f"Saved: {filename} — {start_time} to {end_time} sec — Speaker {speaker}")
```

```python
import os
import shutil
import whisper

model_name = "medium"
cache_path = os.path.expanduser(f"~/.cache/whisper/{model_name}.pt")
custom_model_dir = "./whisper_model_dir"
custom_model_path = os.path.join(custom_model_dir, f"{model_name}.pt")

# 1. Download model if not already cached
if not os.path.exists(cache_path):
    print(f"Downloading {model_name} model...")
    model = whisper.load_model(model_name)
else:
    print(f"{model_name}.pt already cached.")

# 2. Copy to custom folder
os.makedirs(custom_model_dir, exist_ok=True)

if not os.path.exists(custom_model_path):
    shutil.copy(cache_path, custom_model_path)
    print(f"Model copied to: {custom_model_path}")
else:
    print("Model already exists in custom folder.")
```

```python
import os
from pydub import AudioSegment
from docx import Document

output_docx = "transcription.docx"

def transcribe_chunk(volume_boost_db=10):
      # Create a new Word document
    if not os.path.exists(output_docx):
        doc = Document()
        doc.add_heading("Speaker Transcriptions", 0)
    else:
        doc = Document(output_docx)  # Open existing if needed

    for entry in ls[:]:  # copy to avoid modifying list
        speaker = entry["speaker"]
        filename = entry["filename"]
        start = entry["start"]
        end = entry["end"]

        # Load and boost
        audio = AudioSegment.from_wav(filename)
        boosted_audio = audio + volume_boost_db
        boosted_path = f"boosted_{os.path.basename(filename)}"
        boosted_audio.export(boosted_path, format="wav")

        # Transcribe
        result = model.transcribe(boosted_path, language="en")

        # Cleanup
        os.remove(boosted_path)
        ls.remove(entry)

                # Write to docx file
        doc.add_paragraph(f"{speaker}: {start} to {end} sec:\n{result['text']}\n")

    # Save the document
    doc.save(output_docx)
        # Print speaker + duration + transcription
        # print(f"{speaker}: {start} to {end} sec: {result['text']}")
```

```python
import os

def delete_wav_files(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".wav"):
            filepath = os.path.join(directory, filename)
            os.remove(filepath)
            print(f"Deleted: {filepath}")

# Example usage
delete_wav_files("/content")
```
